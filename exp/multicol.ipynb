```python
import sys
import os
import numpy as np

# Add the parent directory to path to import toolkit and models
sys.path.append(os.path.abspath('..'))

from toolkit import ModelMetrics

# Set random seed for reproducibility
np.random.seed(42)

# --- Create a toy dataset ---
n_samples = 100

# 1. Independent Feature
X1 = np.random.randn(n_samples)          

# 2. Independent Feature
X2 = np.random.randn(n_samples)          

# 3. Highly Correlated Feature (Depends heavily on X1 and X2, but has some noise)
X3 = 2.0 * X1 + 3.0 * X2 + np.random.randn(n_samples) * 0.1  

# 4. Perfectly Collinear Feature (Exact multiple of X1)
X4 = -5.0 * X1                           

# Combine into our "features" list expected by the metric function
features = [X1, X2, X3, X4]

# --- Test the Multicollinearity Metric ---
# Using the new static method approach!
vif_scores = ModelMetrics.multicollinearity_metric(features)

print("VIF for X1 (Independent with X2, but X3/X4 depend on it):", round(vif_scores[0], 2))
print("VIF for X2 (Independent with X1, but X3 depends on it):", round(vif_scores[1], 2))
print("VIF for X3 (Highly Correlated with X1 and X2):", round(vif_scores[2], 2))
print("VIF for X4 (Perfectly Collinear with X1):", vif_scores[3])
```
